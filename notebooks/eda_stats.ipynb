```python
import pandas as pd

df = pd.read_csv("../data/churn_data.csv")

df.head()
```

```python
churned = df[df['Churn Value'] == 1]
not_churned = df[df['Churn Value'] == 0]
```

```python
churned['Tenure Months'].describe()
```

```python
not_churned['Tenure Months'].describe()
```

```python
df.groupby('Churn Value')['Tenure Months'].mean()
```

```python
churned['Monthly Charges'].describe()
```

```python
not_churned['Monthly Charges'].describe()
```

```python
df.groupby('Churn Value')['Monthly Charges'].mean()
```

```python
df.groupby('Contract')['Churn Value'].mean().sort_values(ascending=False)
```

Contract type has a strong impact on churn. Customers with month-to-month contracts show the highest churn rate (42.7%), while customers with two-year contracts have the lowest churn rate (2.8%).
This indicates that customers with long-term commitments are significantly more likely to remain with the company.

```python
df.groupby('Internet Service')['Churn Value'].mean().sort_values(ascending=False)
```

```python
df.groupby('Payment Method')['Churn Value'].mean().sort_values(ascending=False)
```

```python
df.groupby('Internet Service')['Churn Value'].mean()
```

```python
df.groupby('Payment Method')['Churn Value'].mean()
```

Payment method significantly affects churn behavior. Customers using electronic check show the highest churn rate (45.3%), while customers using automatic payments such as credit cards or bank transfers show much lower churn rates (~15–17%).
This suggests that customers enrolled in automatic payment systems are more likely to stay with the service.

```python
import matplotlib.pyplot as plt

plt.hist(churned['Tenure Months'], bins=30, alpha=0.6, label='Churned')
plt.hist(not_churned['Tenure Months'], bins=30, alpha=0.6, label='Not Churned')

plt.xlabel("Tenure Months")
plt.ylabel("Customers")
plt.title("Tenure Distribution by Churn")
plt.legend()
plt.show()
```

```python
import seaborn as sns

sns.countplot(x='Contract', hue='Churn Value', data=df)
plt.title("Churn by Contract Type")
plt.show()
```

```python
sns.countplot(x='Payment Method', hue='Churn Value', data=df)
plt.xticks(rotation=45)
plt.title("Churn by Payment Method")
plt.show()
```

Customers with lower tenure churn more frequently
• Higher monthly charges correlate with higher churn
• Month-to-month contracts show the highest churn rate (~43%)
• Fiber optic users churn significantly more than DSL users
• Electronic check payment method has the highest churn rate (45%)
